```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set_style('black')

def factorial(x):
    n = 1
    while x != 0:
        n = n * x
        x = x - 1
    return n

def combination(n, k):
    c = factorial(n) / (factorial(n - k) * factorial(k))
    return c

def binomial_distribution(n, k, p):
    """
        Here, 
        n: The given number of trials
        k: Number of Successes
        p: Probability of success on a given trial
    """
    likelihood = combination(n, k) * p**k * (1 - p)**(n - k)
    return round(likelihood, 4)

binomial_distribution(10, 1, 0.5)

def distrubution_data(n, p):
    points = {}
    values = [binomial_distribution(n, k, p) for k in range(n+1)]
    keys = [k for k in range(n+1)]
    points['Keys'] = keys
    points['Value'] = values
    return points

def add_text(data):
    for i in range(len(data['Keys'])):
        y_coor = data['Value']
        plt.text(i, y_coor[i], round(y_coor[i], 3), 
                 ha='center', va='bottom', 
                 color='black')
    
def distribution_graph(n, p):
    data = distrubution_data(n, p)
    plt.figure(figsize=(10,5))
    ax = sns.barplot(data=data, x='Keys', y='Value',
                edgecolor='yellow',
                width=0.8,
                color='coral')    
    ax.set_facecolor('#c9daf8')
    add_text(data)
    plt.show()

# n = int(input("Enter the total number of Trials:"))
# p = float(input('Enter the probabilty of success of each trial:'))
distribution_graph(10, 0.167)
```

```python
import matplotlib.pyplot as plt

def add_custom_labels(categories, values, custom_labels):
    # Create a bar chart
    plt.bar(categories, values)

    # Add custom text labels to the bars
    for i in range(len(custom_labels)):
        plt.text(i, values[i], values[i], ha='center', va='bottom')

    # Customize the chart further (if needed)
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.title('Bar Chart with Custom Labels')
    plt.grid(True)

    # Show the chart
    plt.show()

# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 45, 60]
custom_labels = ['Label A', 'Label B', 'Label C', 'Label D']

# Call the function to create the bar chart with custom labels
add_custom_labels(categories, values, custom_labels)
```

```python
covid_df.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace=True, axis=1)
```

```python
covid_df
```

```python
# Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=['Cured', 'Deaths', 'Confirmed'],
                          index= 'State/UnionTerritory', aggfunc=max)
```

```python
statewise.head()
```

```python
statewise['Recovery_rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality_rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values('Confirmed', ascending=False)
```

```python
statewise.style.background_gradient(cmap='inferno')
```

```python
top10_active = covid_df.groupby("State/UnionTerritory").max()[["Active_Cases", "Date"]]. \
    sort_values("Active_Cases", ascending=False).reset_index().head(10)
```

```python
plt.figure(figsize=(15,7))
ax = sns.barplot(data=top10_active, x='State/UnionTerritory', y='Active_Cases', linewidth=1, edgecolor='black')
plt.title('Top 10 Active Cases States', size=20)
# plt.xticks(rotation=50)
plt.show()
```

```python
top10_deaths = covid_df.groupby("State/UnionTerritory").max()[["Deaths", "Date"]]. \
    sort_values("Deaths", ascending=False).reset_index().head(10)
```

```python
plt.figure(figsize=(15,7))
ax = sns.barplot(data=top10_deaths, x='State/UnionTerritory', y='Deaths', linewidth=1, edgecolor='black')
plt.title('Top 10 Deaths by States', size=20)
# plt.xticks(rotation=50)
plt.show()
```

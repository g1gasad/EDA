```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('E:\\Raw Data\\Spotify\\tracks.csv')
df_tracks.head()
```

```python
# Null values
pd.isnull(df_tracks).sum()
# returns the sum of null values in each of the columns
```

```python
df_tracks.info()
```

```python
# sorting the df according to popularity
sorted_df = df_tracks.sort_values('popularity', ascending=True).head()
sorted_df
```

```python
df_tracks.describe().T
```

```python
# df_tracks[df_tracks['popularity'] > 90]
# df_tracks.query('popularity > 90')
most_popular = df_tracks.loc[df_tracks['popularity'] > 90] \
    .sort_values('popularity', ascending=False)
most_popular.head(10)
```

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks
```

```python
df_tracks[['artists']].iloc[10]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'], axis=1).corr(method="pearson")
```

```python
plt.figure(figsize=(12,5))
heatmap = sns.heatmap(corr_df, annot=True, cmap='inferno', linewidths=0.5, linecolor='cyan')
plt.title("Correlation Heatmap between Variables")
plt.show()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
# returns 4% of the total data
```

```python
sample_df.shape
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data= sample_df, x= "energy", y= "loudness", color='c')
plt.title('Energy VS Loudness Correlation')
plt.xlabel("Energy")
plt.ylabel('Loudness')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data= sample_df, x= "acousticness", y= "popularity", color='b')
plt.title('Acousticness VS Popularity Correlation')
plt.xlabel("Acousticness")
plt.ylabel('Popularity')
plt.show()
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
# instead of release_date, we can even use 0 since there are no multiple levels
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
year = df_tracks['dates'].dt.year
```

```python
sns.displot(year, discrete=True, kind='hist', height=5, aspect=2) \
    .set(title="Number of Songs per Year")
plt.show()
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x= year, y= total_dr, ax=ax, errwidth=False) \
    .set(title='Duration of songs by year')
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr = df_tracks.duration
fig_dims = (10,5)
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.lineplot(x= year, y= total_dr, ax=ax) \
    .set(title='Duration of songs by year')
plt.xticks(rotation=50)
plt.show()
```

```python
df_genre = pd.read_csv('E:\\Raw Data\\Spotify\\SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.figure(figsize=(13,5))
plt.title("Duration of songs in different genres")
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre, errwidth=False)
plt.xlabel('Duration in milli-seconds')
plt.ylabel('Genres')
```

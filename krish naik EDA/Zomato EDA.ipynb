```python
# we will try to combine the two datasets such that we can get the name of country instead of a number
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
# or final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
# Pie Chart
# Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

Observations:
Most of the transactions or records are from India followed by USA and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size() \
            .reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color',
            palette=['black', 'red', 'orange', 'yellow', 'green', 'green'], width=3)
plt.show()
```

Observation:
    
    1. Not Rated count is very high
    2. Maximum number of ratings are between 2.5 anf 4.0

```python
## Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])

# this is not the rating count but the frequency of their appearance in the records
```
